```python
df_raw = pd.read_csv('insu_pre_review.csv',encoding = 'euckr')
```

```python
join_1 = pd.read_csv('join_1.csv')
join_1 = join_1[['customer_id','risk']] # 1분류 목표변수

join_2 = pd.read_csv('join_2.csv')
join_2 = join_2[['customer_id','n_risk']] # 2분류 목표변수
```

```python
df_1 = pd.merge(df_raw,join_1,on='customer_id',how = 'left') # 1분류 
df_2 = pd.merge(df_raw,join_2,on='customer_id',how = 'left') # 2분류
```

```python
review_1 = df_1[df_1['risk'].isna() == False]
review_2 = df_2[df_2['n_risk'].isna() == False]
```

```python
# 1분류 train용 join data 내보내기
review_1.to_csv('review_1.csv',encoding = 'euckr')

# 2분류 train용 join data 내보내기
review_2.to_csv('review_2.csv',encoding = 'euckr')
```
