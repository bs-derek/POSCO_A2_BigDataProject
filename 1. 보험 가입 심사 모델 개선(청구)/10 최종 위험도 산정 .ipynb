```python
pay_1 = pd.read_excel('normal_final_classification.xlsx',encoding='euckr')
pay_2 = pd.read_excel('blood_final_classification.xlsx',encoding='euckr')
```

```python
req = pd.read_csv('req_predict.csv',encoding='euckr')
```

```python
req = req[['customer_id','risk']]
```

```python
pay = pd.concat([pay_1,pay_2])[['customer_id','final_risk']]
```

```python
d1 = pd.merge(req,pay,on='customer_id',how='inner')
```

```python
d1['last_judge'] = d1['risk'] + d1['final_risk']
```

```python
d2 = pd.merge(req,pay,on='customer_id',how='left')
d2 = d2[d2['final_risk'].isna() == True]
```

```python
d2['final_risk'] = 0
d2
```

```python
d2['final_risk'] = np.where(d2['risk']==1,d2['risk']+1,d2['final_risk'])
d2['final_risk'] = np.where(d2['risk']==2,d2['risk']+2,d2['final_risk'])
d2.groupby('final_risk').count()
```

```python
d2['last_judge'] = d2['final_risk']
d2
```

```python
d1
```

```python
judge_reject = pd.concat([d1,d2]).drop(['risk','final_risk'],axis=1)
```
