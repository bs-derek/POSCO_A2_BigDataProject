```python
g_high=pd.read_csv('high_danger_g.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
g_low=pd.read_csv('low_danger_g.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
g_profit=pd.read_csv('profit_reject_g.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
```

```python
b_high=pd.read_csv('high_danger.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
b_low=pd.read_csv('low_danger.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
b_profit=pd.read_csv('profit_reject.csv',encoding='euckr').drop('Unnamed: 0',axis=1)
```

```python
high = pd.concat([g_high,b_high])
low = pd.concat([g_low,b_low])
profit = pd.concat([g_profit,b_profit])
```

```python
low['risk'] = 1
```

```python
high['risk'] = 2
```

```python
df = pd.concat([high,low,profit])
df.groupby('risk').count()
```

```python
df_raw = pd.read_csv('insu_pre_review.csv',encoding = 'euckr')
df_pro = pd.merge(df_raw,df,on='customer_id',how='left')
df_pro = df_pro[df_pro['risk'].isna() == False]
df_pro
```
