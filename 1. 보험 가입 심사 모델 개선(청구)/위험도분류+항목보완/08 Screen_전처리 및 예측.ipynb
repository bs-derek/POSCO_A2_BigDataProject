```python
df = pd.read_csv('insu_nh_h_screen.csv',encoding='euckr')
```

```python
np.random.seed(123)
sample = df.sample(frac=0.2)
```

```python
test = sample.reset_index()
test.to_csv('raw_screen.csv',encoding='euckr')
```

```python
data = sample[['bp_max','bp_min','bp_gluc','bt_mch','bt_chol','gender','smoke_flag','drinking_flag']]
ls = ['bp_max','bp_min','bp_gluc','bt_mch','bt_chol','gender','smoke_flag','drinking_flag']
```

```python
for i in ls :
    data = data[data[i] != '.']
data
```

```python
data_2 = data.astype('float')
```

```python
data_2.info()
```

## 1분류 유의변수

'bt_chol','bp_judge','bt_gluc_judge','bt_mchc_judge','gender'

- bt_gluc : 100 미만 정상, 100~125 당뇨 전 단계
- mch : 남자 14-17.5 g/dL, 여자 12.3-15.3 g/dL
- 성별 : 남자 1 여자 2
- 혈압 : max 120 130mmHg, min 80 85

```python
data_2['bp_judge'] = np.where((data_2['bp_max'].between(0,130)) & (data_2['bp_min'].between(0,85)),0,1)
data_2['bt_gluc_judge'] = np.where(data_2['bp_gluc']<100,0,1)
```

```python
data_2.groupby('bt_gluc_judge').count()
```

```python
data_m = data_2[data_2['gender'] == 1]
data_f = data_2[data_2['gender'] == 2]
data_m['bt_mchc_judge'] = np.where(data_m['bt_mch'].between(14,17.5),0,1)
data_f['bt_mchc_judge'] = np.where(data_f['bt_mch'].between(12.3,15.3),0,1)

data_3 = pd.concat([data_m, data_f])
data_3
```

```python
data_3.groupby('bt_mchc_judge').count()
```

```python
data_3 = data_3.reset_index()
data_3
```

```python
# 1분류 스크린 데이터 내보내기_35000개 데이터 샘플링
data_3.to_csv('screen_1.csv',encoding='euckr')
```
